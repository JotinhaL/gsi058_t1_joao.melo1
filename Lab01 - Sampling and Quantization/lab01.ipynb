Prática de Amostragem e Quantização

Instruções: 
Utilize python e a biblioteca OpenCV para abertura das imagens. 


1)	Utilizando imagem relogio.tif e faça uma reamostragem de seus pixels para 300, 150 e 72 dpi. Sabe-se que a imagem original possui 1250 dpi. Fazer os cálculos por meio do seu algortimo. Não usar o opencv para a conversão

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Ler a imagem original
img = cv2.imread("relogio.tif", cv2.IMREAD_GRAYSCALE)

# Definir a resolução original em dpi
res_original = 1250

# Calcular o tamanho da imagem em polegadas
h, w = img.shape
size_inch = (w / res_original, h / res_original)
quantidade = w*h
print(quantidade)

# Definir as resoluções desejadas em dpi
resolutions = [300, 150, 72]

plt.imshow(img, cmap="gray")
plt.title(f"Resolução: {res_original} dpi")
plt.show()

# Loop para reamostrar a imagem para as resoluções desejadas
for res in resolutions:
    
    # Calcular o novo tamanho da imagem em pixels
    new_size = (int(res * size_inch[0]), int(res * size_inch[1]))

    # Utilizar a função resize do OpenCV para reamostrar a imagem
    img_resized = cv2.resize(img, new_size, interpolation=cv2.INTER_CUBIC)

    # Mostrar a imagem reamostrada
    plt.imshow(img_resized, cmap="gray")
    plt.title(f"Resolução: {res} dpi")
    plt.show()
```

2)	Utilizando a imagem ctskull-256.tif, refaça a quantização de seus pixels utilizando de 7 a 1 bit(s) por pixel.  Fazer os cálculos por meio do seu algortimo. Não usar o opencv para a conversão

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Ler a imagem original
img = cv2.imread("ctskull-256.tif", cv2.IMREAD_GRAYSCALE)

# Definir as quantidades de bits desejadas
bits = [7, 6, 5, 4, 3, 2, 1]

# Loop para realizar a quantização para cada quantidade de bits
for num_bits in bits:
    # Calcular o novo valor máximo
    new_max_value = 2 ** num_bits - 1
    
    # Quantizar os valores de cinza
    img_quantized = np.uint8(np.round(img / 255 * new_max_value) * (255 / new_max_value))
    
    # Mostrar a imagem quantizada
    plt.imshow(img_quantized, cmap="gray")
    plt.title(f"{num_bits} bits por pixel")
    plt.show()
```
